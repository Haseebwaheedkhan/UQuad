```python
# Reading Files
df_passage = pd.read_excel ("NLP Project Dataset .xlsx",sheet_name=0)
df_qa = pd.read_excel ("NLP Project Dataset .xlsx",sheet_name=1)
```

```python
def QuestionAnswer(data_frame, data_frame_qa, passage_Id):
    '''a
    Input DataFrame and paragraph number whose Question and answer are to be extracted
    Return Question and answers in form of dictionary
    '''
    for index in range(len(data_frame)):
        
        # Check Passage ID with the Passage ID Of complete dataset
        if data_frame["PassageId"][index] == passage_Id: 
            questionid = 0
            answer = {}
            question = {}
            temp = {}

            for index in range(len(data_frame_qa)):
                if data_frame_qa["PassageId"][index] == passage_Id:
                    

                    while data_frame_qa["PassageId"][index] == passage_Id:

                        
                        temp = {}
                        temp_answer = {}

                        temp["question"] = data_frame_qa["Question"][index]
                        temp_answer[0] = data_frame_qa["Ans_1"][index]
                        temp_answer[1] = data_frame_qa["Ans_2"][index]
                        temp_answer[2] = data_frame_qa["Ans_3"][index]
                        temp["answer"] = temp_answer
                        
                        question[questionid] = temp
                        index +=1
                        questionid +=1
                        
                        # Stop looping further if last row found
                        if(index == len(data_frame_qa)):

                            break
                    break
            return question
```

```python
#Creating Json file of Urdu Dataset
import json as js

json = {}
for i in range(len(df_passage)):
    temp = {}
    temp["question"] = QuestionAnswer(df_passage, df_qa, df_passage["PassageId"][i])
    temp["context"] = df_passage["Text"][i]
    json[str(df_passage["PassageId"][i])] = temp


with open('UQuAd.json', 'w') as json_file:
    js.dump(json, json_file)
```

```python
# Loading Json File 

with open('UQuAd.json', "r", encoding="utf-8") as file:
    UQuAd = js.load(file)
```
